```python
import random

simbolos = ["🍉", "🍊", "🍌", "🍀"]

nom = input("Digite su nombre: ")

class Player():
  def __init__(self, nombre):
    self.nombre = nombre
    self.monedas = 100

  def apostar(self, simbolos):
    self.monedas -= 10
    print("-10 monedas")
    machine = []
    n = 0
    num = 0
    dol = 0
    por = 0
    y = 0
    while n < 3:
      n +=1
      k = random.choice(simbolos)
      machine.append(k)
    print(f"[{machine[0]}] [{machine[1]}] [{machine[2]}]")
    for i in machine:
      if i == "🍉":
        num += 1
      elif i == "🍊":
        dol += 1
      elif i == "🍌":
        por += 1
      else:
        y += 1
    cantidad = [num, dol, por, y]
    repetidos = max(cantidad)
    if repetidos == 2:
      self.monedas += 20
      return [2, 20]
    if repetidos == 3:
      self.monedas += 50
      return [3, 50]
    return [0, 0]

  def mostrar_saldo(self):
    print(f"Monedas: {self.monedas}")

jugador1 = Player(nom)

while True:
  jugador1.mostrar_saldo()
  t = input("¿Tirar? (s/n): ")
  while t != "s" and t != "n":
    t = input("¿Tirar? (s/n): ")

  if t == "s":
    m = jugador1.apostar(simbolos)
    print(f"¡{m[0]} iguales! Ganas {m[1]} monedas")
  else:
    print()
    jugador1.mostrar_saldo()
    print(f"Gracias por jugar, {jugador1.nombre}")
    break
  print()
```
